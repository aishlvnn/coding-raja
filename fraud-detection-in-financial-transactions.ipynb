## FRAUD DETECTION IN FINANCIAL TRANSACTIONS

dataset link : https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score
from sklearn.metrics import precision_recall_fscore_support as score
```

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
input1 = df.iloc[0, :30].values
input1
```

```python
df.columns
```

```python
df.info()
```

```python
df["Class"].value_counts()
```

Number of legit transactions : 284315, Number of illegal transactions : 492

```python
legit_df = df[df["Class"] == 0]
illegal_df = df[df["Class"] == 1]
legit_df.head()
```

```python
illegal_df.head()
```

```python
df.groupby("Class").mean()
```

```python
legit_rnd_df = legit_df.sample(1500)
legit_rnd_df.shape
```

```python
new_df = pd.concat([legit_rnd_df, illegal_df])
new_df.groupby("Class").mean()
```

```python
x = new_df.iloc[:, :30]
x.head()
```

```python
y = new_df.iloc[: , 30]
y.head()
```

```python
x = x.values
y = y.values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 5)
```

```python
reg = LogisticRegression()
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
reg.predict([input1])
```

```python
mean_absolute_error(y_pred, y_test)
```

```python
mean_squared_error(y_pred, y_test)
```

```python
r2_score(y_pred, y_test)
```

```python
precision, recall, fscore, support = score(y_pred, y_test)
print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))
print('support: {}'.format(support))
```

```python
accuracy_score(y_pred, y_test)
```
